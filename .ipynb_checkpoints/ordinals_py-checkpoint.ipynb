# Numerals 1-26 to list
Using a list of numerals 1 - 26 to create 26 globals

```python
import pandas as pd
import random
import string
```

```python
dirtyNumerals = """1st – first
2nd – second
3rd – third
4th – fourth
5th – fifth
6th – sixth
7th – seventh
8th – eighth
9th – ninth
10th – tenth
11th – eleventh
12th – twelfth
13th – thirteenth
14th – fourteenth
15th – fifteenth
16th – sixteenth
17th – seventeenth
18th – eighteenth
19th – nineteenth
20th – twentieth
21st – twenty-first
22nd – twenty-second
23rd – twenty-third
24th – twenty-fourth
25th – twenty-fifth
26th – twenty-sixth"""
```

```python
#Cleaning dirtyNumerals
#making oridinals list for globals(i)
removeChars = ['–', '\n']
for i in removeChars:
    dirtyNumerals = dirtyNumerals.replace(i, ' ')
dirtyNumerals = dirtyNumerals.split(" ")
#dirtyNumerals
```

```python
emptyIdx = []
i = 0
for j in dirtyNumerals:
    if j == '':
        emptyIdx.append(dirtyNumerals.index('', i))
        i += 1
    else:
        i += 1
```

```python
#prints all elements inline with a comma
def printInline(toPrint):
    for i in toPrint:
        print(str(i) + ', ', end = " ")
```

```python
printInline(emptyIdx)
```

```python
def removeSpaces(listName):
    filterLambda = filter(lambda x: x != '', listName)
    cleanList = list(filterLambda)
    return cleanList
```

```python
cleanNumerals = removeSpaces(dirtyNumerals)
print(cleanNumerals)
```

```python
#Globals[]i can not use variables that start with nums
#Setting up string list of numbers to match and remove to get word numerals list
strNums = []
for i in range(10):
    strNums.append(str(i))
```

```python
#Creating a function to get the matches
#Remove matching values and returns a tuple of two lists, one of matches, one unmatched
def getMatching(list1, list2):
    list2 = tuple(list2)
    matches = [item for item in list1 if item.startswith(list2)]
    unmatched = list(list1)
    #print(matches)
    #print(unmatched)
    n = 0
    while n < len(unmatched):
        for match in matches:
            if unmatched[n] == match:
                unmatched.remove(unmatched[n])
                n+=1
            else:
                n+=1
    return matches, unmatched
```

```python
#Replacing - with _ becase variables dont like - 
twoReturns = (getMatching(cleanNumerals, strNums))
# print(twoReturns[1])
ord_names1 = []
for i in twoReturns[1]:
    ord_names1.append(i.replace('-', '_'))
ord_names1
```

```python
#Creates 36^6=2 176 782 336  different combinations of letters and numbers
def create_random_str(size= 6, chars=string.ascii_uppercase + string.ascii_lowercase + string.digits + string.punctuation):
    return ''.join(random.choice(chars) for _ in range(size))
```

```python
values = []
for i in range(0, 26):
    j = random.randint(4, 12)
    values.append(create_random_str(j))
    print(create_random_str(j))
    
temp_dict = dict(zip(ord_names1,values))
temp_dict
```

# Word Freq
['Notre Dam Cryptography']('https://www3.nd.edu/~busiforc/handouts/cryptography/letterfrequencies.html')
- link above is reference to percentages, used below

```python
def insertRow(rowNum, df, rowVal):
    df1 = df[0:rowNum]
    df2 = df[rowNum:]
    df1.loc[rowNum] = rowVal
    dfResult = pd.concat([df1, df2])
    #dfResult = dfResult.index
#     dfResult = [*range(dfResult).shape[0]]
    return dfResult
```

```python
# word_frequency = pd.read_clipboard()
# word_frequency.to_csv('word_freq.csv', index=False)
word_freq= pd.read_csv('word_freq.csv')
word_freq
row1 = []
for col in word_freq.columns:
    row1.append(col)
cols_list = ['letter1', '%_freq1', 'commonality1', 'letter2', '%_freq2', 'commonality2']
for i in range(len(row1)):
    word_freq.rename(columns={word_freq.columns[i]: cols_list[i]}, inplace = True)
word_freq = (insertRow(0, word_freq, row1)).reset_index(drop=True)
word_freq
# word_frequency = word_frequency.drop(word_frequency.columns[1], axis=1)
word_freq1 = word_freq[['letter1', '%_freq1', 'commonality1']]
word_freq2 = word_freq[['letter2', '%_freq2', 'commonality2']]
cols_list = ['letter', '%_freq', 'commonality']
for i in range(len(word_freq1.columns)):
    word_freq1.rename(columns={word_freq1.columns[i]: cols_list[i]}, inplace = True)
    word_freq2.rename(columns={word_freq2.columns[i]: cols_list[i]}, inplace = True)
frames = [word_freq1, word_freq2]
word_frequency = (pd.concat(frames)).reset_index(drop = True)
#word_frequency
```

```python
word_frequency['freq_flt'] = [float(i[:-1]) for i in word_frequency['%_freq']]
#word_frequency
```

```python
word_frequency['commonality'][25] = 0.99
```

```python
def is_float(str):
    if type(str) == type(1.1):
        return False
    else:
        return True
word_frequency['common_flt'] = [float(i.strip()) if is_float(i) else i for i in word_frequency['commonality']]
```

```python
word_frequency.to_csv('wrd_freq.csv', index=False)
word_frequency
```

```python
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
plt.style.use('seaborn')
plt.figure(figsize=(20,8))
N = len(word_frequency)
x = word_frequency['letter']
y = word_frequency['freq_flt']
y2 = word_frequency['common_flt']

ind = np.arange(N)
width=0.80
plt.bar(ind, y, width, label='Freq_%')
plt.bar(ind, y2, width, bottom=y, label='Commonality')
plt.ylabel('Value')
plt.title('Letter Frequency and Commonality')
x_pos = [i for i, _ in enumerate(x)]
plt.xlabel('Letter')
plt.xticks(ind + width/2, (x))
plt.legend(loc='best')
plt.show()
```

# Find Ordinals
- by num location

```python
def getOrdinalsList(chrType, begin, end):
    ordList = []
    for i in range(begin, end):
        ordList.append(i)
    #print(ordList)
    #print(type(ordList[0]))
#     print(f'The {chrType} ASCII begins with the number: {ordList[-len(ordList)]},'\
#           + f' and the {chrType}: {chr(ordList[-len(ordList)])}')
#     print(f'Negative len: {-len(ordList)}')
#     print(f'End of list idx: {len(ordList)-1}')
#     print(f'The {chrType} ASCII ends with the number: {ordList[len(ordList)-1]},'\
#           + f' and the {chrType}: {chr(ordList[len(ordList)-1])}')
    return ordList
```

```python
#This is the ascii location of all the symbols, spread out between other ordinals
symbolNumDict = {
    '33':48,
    '58':65,
    '91':97,
    '123':126
}
```

```python
#Patching together all the Ordinal Numeral Locations of Symbols
symbolList2 = []
for key, value in symbolNumDict.items():
    symbolList2.append(getOrdinalsList('Symbol', int(key), value))
symbolList2
symbolList = []
for i in symbolList2:
    for j in i:
        symbolList.append(j)
symbol_char_list = []
for i in symbolList:
    symbol_char_list.append(chr(i))
symbol_dict = {}
for i in  symbol_char_list:
    symbol_dict[f'_{ord(i)}'] = i
symbol_dict
symbol_df = (pd.DataFrame.from_dict(symbol_dict, orient='index')).reset_index(drop=True)
symbol_dict
loc = []
for key, value in symbol_dict.items():
    loc.append(key)
symbol_df.insert(0, 'symb_loc', loc)
```

```python
loc = []
upper = []
for key, value in upper_dict.items():
    loc.append(key)
    upper.append(value)
for i in range(5):
    upper.append(np.NaN)
    loc.append(np.NaN) 
alpha_df = symbol_df
alpha_df['upper_loc'] = loc
alpha_df['upper'] = upper
```

```python
loc = []
lower = []
for key, value in lower_dict.items():
    loc.append(key)
    lower.append(value)  
for i in range(5):
    lower.append(np.NaN)
    loc.append(np.NaN)
alpha_df['lower_loc']=loc
alpha_df['lower']= lower
# word_freq2.rename(columns={word_freq2.columns[i]: cols_list[i]}, inplace = True)
alpha_df.rename(columns= {alpha_df.columns[1]: 'symb'}, inplace = True)
alpha_df = alpha_df[['upper', 'upper_loc', 'lower', 'lower_loc', 'symb', 'symb_loc']]
alpha_df.to_csv('alpha_ords.csv', index=False)
alpha_df
```
