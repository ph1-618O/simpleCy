```python
def change_to_int(data):
    return data/100
```

```python
def insertRow(rowNum, df, rowVal):
    df1 = df[0:rowNum]
    df2 = df[rowNum:]
    df1.loc[rowNum] = rowVal
    dfResult = pd.concat([df1, df2])
    #dfResult = dfResult.index
#     dfResult = [*range(dfResult).shape[0]]
    return dfResult
```

```python
# word_frequency = pd.read_clipboard()
# word_frequency.to_csv('word_freq.csv', index=False)
word_freq= pd.read_csv('word_freq.csv')
word_freq
row1 = []
for col in word_freq.columns:
    row1.append(col)
cols_list = ['letter1', '%_freq1', 'commonality1', 'letter2', '%_freq2', 'commonality2']
for i in range(len(row1)):
    word_freq.rename(columns={word_freq.columns[i]: cols_list[i]}, inplace = True)
word_freq = (insertRow(0, word_freq, row1)).reset_index(drop=True)
word_freq
# word_frequency = word_frequency.drop(word_frequency.columns[1], axis=1)
word_freq1 = word_freq[['letter1', '%_freq1', 'commonality1']]
word_freq2 = word_freq[['letter2', '%_freq2', 'commonality2']]
cols_list = ['letter', '%_freq', 'commonality']
for i in range(len(word_freq1.columns)):
    word_freq1.rename(columns={word_freq1.columns[i]: cols_list[i]}, inplace = True)
    word_freq2.rename(columns={word_freq2.columns[i]: cols_list[i]}, inplace = True)
frames = [word_freq1, word_freq2]
word_frequency = (pd.concat(frames)).reset_index(drop = True)
#word_frequency
```

```python
word_frequency['freq_flt'] = [float(i[:-1]) for i in word_frequency['%_freq']]
#word_frequency
```

```python
word_frequency['commonality'][25] = 0.99
```

```python
def is_float(str):
    if type(str) == type(1.1):
        return False
    else:
        return True
word_frequency['common_flt'] = [float(i.strip()) if is_float(i) else i for i in word_frequency['commonality']]
```

```python
#Function expects string
def most_common(text):
    d = collections.defaultdict(int)
    for c in text:
        d[c] += 1
    common = sorted(d.items(), key=lambda x: x[1], reverse=True)[0]
    return common
```

```python
def drop_letter(letter, test_text):
    #print(letter)
    #print(test_text)
    return test_text.replace(letter, "")   
    
```

```python
def tuple_frequency(test_text):
    # Iterator through input
    # While loop counts frequency of letters and removes them
    counter = len(test_text)
    drop_text = test_text
    lst_common = []
    while counter != 0:
        key_finder = most_common(drop_text)
        lst_common.append(key_finder)
        drop_text = drop_letter(key_finder[0], drop_text)
        if len(drop_text) == 0:
            return lst_common
            break
        counter = counter - 1
```
